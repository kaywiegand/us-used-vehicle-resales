## Scenario

Du arbeitest als Data Scientist für einen US-amerikanischen Gebrauchtwagenhändler. Der Händler kauft gebrauchte Autos günstig in Onlineauktionen und von anderen Autoverkäufern, um sie dann auf der eigenen Plattform gewinnbringend weiterzuverkaufen. Es ist nicht immer einfach zu erkennen, ob sich der Kauf eines Gebrauchtwagens lohnt: Eine der größten Herausforderungen bei Gebrauchtwagenauktionen ist das Risiko, dass ein Auto solch schwerwiegende Probleme hat, die verhindern, dass es an Kunden weiterverkauft werden kann. Hierbei handelt es sich um sogenannte "Montagsautos" - also Autos, die von Hause aus erhebliche Mängel aufgrund von Produktionsfehlern aufweisen, welche die Sicherheit, die Verwendung oder den Wert dieses Autos erheblich beeinträchtigen und gleichzeitig nicht in einer angemessenen Anzahl an Reparaturen oder innerhalb eines bestimmten Zeitraums behoben werden können. Der Kunde hat in diesem Fall das Recht, sich den Kaufpreis zurückerstatten zu lassen. Neben den Anschaffungskosten führt der Fehleinkauf von solchen "Montagsautos" also zu erheblichen Folgekosten, wie z.B. der Einlagerung und Reparatur des Wagens, welche in Verlusten beim Weiterverkauf des Fahrzeugs resultieren können.

Deshalb ist es deiner Chefin wichtig, möglichst viele Fehlkäufe von "Montagsautos" auszuschließen. Um die Einkäufer im Unternehmen bei der riesigen Anzahl an Angeboten zu entlasten, sollst du ein Modell entwickeln, welches vorhersagt, ob ein Angebot ein Fehlkauf im Sinne eines Montagsautos wäre. Das darf allerdings nicht dazu führen, dass zu viele gute Käufe ausgeschlossen werden. Genauere Angaben zu den Kosten und Gewinnen der jeweiligen Käufe erhältst du für die Entwicklung des Prototyps noch nicht.

* **Transparenz:** Wir können jederzeit `df_eda` (vorher) mit `df_clean` (nachher) vergleichen, um den Datenverlust und Qualitätsgewinn zu validieren.

```text
data/
├── 01_raw/                 # Source of Truth (Unverändert)
│   ├── data_train.csv      # Die initiale Rohdatei
│   └── features_aim.csv    # Ziel-Definitionen
│
├── 02_interim/             # Zwischenstände (Split)
│   ├── X_train.parquet     # Roh-Split ohne Engineering
│   ├── X_test.parquet      
│   ├── y_train.parquet     
│   └── y_test.parquet      
│
├── 03_processed/              # Finaler Input für Modeling (Ready)
│   ├── X_train_final.parquet  # Inklusive aller feat_ Spalten
│   └── y_train.parquet        # Zugehöriges Target
│
└── 04_models/                     # Output der Experimente
    ├── model_results.csv          # Dein CSV-Tracker Backup
    └── [Sprechender_Name].joblib  # Gespeicherte Modelle

## Define Metric

Es handelt sich um eine **Klassifizierungsaufgaben**, Montagsauto ja/nein. Entsprechend dem Datenwörterbuch IsBadBuy 1/0.   
Die primäre Metrik **F1-SCORE** ist sinnvoll. 👉 Hypothese: Recall wichtiger als Precision → evtl. später Threshold-Tuning


**Erläuterung:** 

Ein Montagsauto bringt nicht nur keinen Gwinn, sondern kann für erhebliche Verluste sorgen. Dem nach ist das Vermeiden eines Montagsauto wichtiger, als die fehlerhafte Identifikation als solches. Die Chefin hat explizit gesagt: "Möglichst viele Fehlkäufe ausschließen".
Das entspricht der Maximierung der Trefferquote: *RECALL*.

Gleichzeit sollten nicht zu viele gute Autos übersehen werden. Klassische Zwickmühle der Balance. Dadurch entgehen dem Unternehmen profitable Möglichkeitn und es entstehen Opportunitätskosten. Die *PRECISION* darf also nicht zu sehr sinken.

Eine Frage der Balance. Der wichtigste Faktor für das Model ist die Maximierung des Recalls bei Einhaltung einer akzeptablen Precision-Untergrenze. Deshalt ist für unsere Projekt der **F1-SCORE** als harmonische Mittel aus Precision und Recall die primäre Metrik.

Interessant könnte auch der F2-Score / F-Beta Wert sein. Der Recall zählt dabei doppelt so viel wie die Precision. Das ist perfekt für diesen Fall, da die Auftraggeberin möglichst viele  Montagsautos vermeiden möchte.

**Standard-Metriken bei Klassifizierung:**
* Accuracy (Genauigkeit)
* Precision (Genauigkeit/Relevanz)
* Recall (Trefferquote/Sensitivität)
* F1-Score (harmonische Mittel aus Precision und Recall)
  
**Confusion Matrix** 
|                      | Vorhersage 0: GOOD | Vorhersage 1: 1 IsBadBuy |
|:----------------------|:----------------------------------|:---------------------------|
| **Realität 0:** | **True Negative (TN)** | **False Positive (FP)** |
| GOOD         | *Korrekt:* Auto ist gut, wir     | *Fehlalarm:* Auto ist gut,  |
|                      | kaufen und machen Profit.        | wir kaufen es fälschlich  |
|                      |                                  | nicht (entgangener Deal). |
| **Realität 1:** | **False Negative (FN)** | **True Positive (TP)** |
| IsBadBuy           | *Risiko:* Wir dachten, es sei gut,| *Korrekt:* Wir haben das   |
|                      | kauften es, aber es ist Schrott. | Montagsauto erkannt und   |
|                      | (Hohe Verluste!)                 | nicht gekauft.            |
